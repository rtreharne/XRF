# The $\chi^2$ Statistic - What is it? Why is it relevant? How do I use it in my work?
## What is it? Why is it relevant?
In a nutshell: the $\chi^2$ statistic is a measure of how likely it is that a certain model describes a set of data. When you divide by the number of degrees of freedom of your model, ideally the answer should be 1. If you learn one thing from this workbook, that should be it. 

Of course, there's more to it than just that. A full breakdown of the $\chi^2$ distribution is beyond the scope of this workbook. Interested students can look here (http://maxwell.ucsc.edu/~drip/133/ch4.pdf) for a more thorough breakdown. We start with the equation for $\chi^2$,

$$ \chi^2 = \sum^{n}_{i=1} \left( \frac{x_i - \mu_i}{\sigma_i} \right)^2 = \sum \left( \frac{Obs - Exp}{\sigma_{Obs}} \right)^2. $$

Looking at the equation, we can see that we have some **observed** experimental values, $x_i$; we have some **expected** values, $\mu_i$, derived from our model; and we have the uncertainty (standard deviation) in the observed values, $\sigma_i$. To answer the starting question: "What is it?", we can now see that the $\chi^2$ statistic is simply the difference between experimental and modelled values, normalised by the uncertainty in the experimental values. From this we can immediately draw two pieces of information:

- a large difference between experimental values and model values will lead to a large $\chi^2$, and
- a small experimental uncertainty will lead to a larger $\chi^2$; a large experimental uncertainty will lead to a small $\chi^2$.

On the latter point - particular care must be made when interpreting a goodness-of-fit $\chi^2$ value: you have a low $\chi^2$ value; is this because your model is good, or but because your experimental uncertainty is very high? 

If a particular $x_i$ is normally distributed around our model value, we expect that it contribute about 1 to our total $\chi^2$ summation. If we have $n$ measurements, and each measurement contributed 1 to the summation, we find,

$$ \chi^2 = \sum^{n} 1n = n. $$

If we then take this sum total and divide it by the number of degrees of freedom in the model, $\nu$, we get,

$$ \chi^2/\nu = \frac{n}{\nu}. $$

Commonly, the number of degrees of freedom is only slightly less than the total number of measurements, $n$. With this approximation, we get the result,

$$ \chi^2/\nu = \frac{n}{\nu} \approx \frac{n}{n} \mathbf{= 1}. $$

*This* is why a $\chi^2/\nu$ (or $\chi^2/$NDF) of 1 is seen as desirable. However, this is not enough on its own to say that the chosen model describes the data well. Inherent assumptions in $\chi^2$ statistics include:

- the assumption that $\chi^2$ is the sum of the squares of random normal deviates, that is, that each $x_i$ is normally distributed about its mean value $µ_i$,
- there is a large sample of data, independently drawn from an even larger sample.

On the first point, it is possible that freak, random outliers can cause a large increase in the $\chi^2$ value. It is commonplace to exclude obvious outliers when calculating this statistic. However, if we have a large $\chi^2$ value that can not be explained by any of the above, we have two options:

- the model represented by the $\mu_i$ is a valid one but that a statistically improbable excursion of $\chi^2$ has occurred, or, 
- that our model is so poorly chosen that an unacceptably large value of $\chi^2$ has resulted.

Conversely, if the $\chi^2$ is very small, we have three options:

- our model is valid but that a statistically improbable excursion of $\chi^2$ has occurred, or,
- we have, over-estimated the values of $\sigma_i$, or,
- someone has given us fraudulent data - that is, data “too good to be true”.

For most of the time in your lab sessions, you will find it uncommon that statistically improbable large $\chi^2$ values will occur - meaning usually the problem is you have data that doesn't fit your model (in the large $\chi^2$ case), or you have over-estimated your errors (very small $\chi^2$ case).

## $\chi^2$ with errors in both x & y variables
The above assumes a regular experiment where one has a dependent (measured) variable with an uncertainty, and an independent (response) variable which does not. In this scenario, the equation for $\chi^2$ looks like this,

$$ \chi^2 = \sum^{n}_{i=1} \left( \frac{y_i - f(x_i)}{\Delta y_i} \right)^2 ;$$

that is to say, we are finding the difference between our measured values and the model values (which are some function of x, usually a straight line), and we only have errors in the y values to divide by.

However, in lab sessions you will commonly measure both the x & y variables, and use these to calculate some sort of gradient or parameter in a curve. How do we calculate the uncertainty $\sigma_i$ now? We know from our error analysis, that a general function of x & y, which we will call $z(x,y)$, has an uncertainty given by:

$$ (\sigma_z)^2 = \left( \frac{\partial Z}{\partial y} \sigma_y \right)^2 + \left( \frac{\partial Z}{\partial x}  \sigma_x \right)^2 .$$

If we have our function, $z(x,y) = y_i - f(x_i)$, then,

$$(\sigma_z)^2 =  \left( \frac{\partial}{\partial y} (y_i - f(x_i)) \sigma_y \right)^2 + \left( \frac{\partial Z}{\partial x} ((y_i - f(x_i)) \sigma_x \right)^2 .$$ 

To evaluate this, we need to define our model function, $f(x_i)$. For simplicity - let's make it a straight line function: $f(x_i) = mx_i + c$. Substituting this into the above equation gives us:

\begin{align}
(\sigma_z)^2 & = \left( \frac{\partial}{\partial y} (y_i - mx_i - c) \sigma_y \right)^2 + \left( \frac{\partial}{\partial x} ((y_i - mx_i - c) \sigma_x \right)^2  \\
& = \left( \mathbf{1} \sigma_y \right)^2 + \left(\mathbf{-m} \sigma_x \right)^2 \\
& = \left( \sigma_y \right)^2 + \left(m \sigma_x \right)^2
\end{align}

Substituting into the equation for $\chi^2$, we get:

$$ \chi^2 = \sum^{n}_{i=1} \left( \frac{y_i - f(x_i)}{\sqrt{\left( \sigma_y \right)^2 + \left(m \sigma_x \right)^2}} \right)^2 $$

This is our equation for $\chi^2$ with errors in both x & y variables. We can see that if we have no error in our x variables $(\sigma_x = 0)$, then the denominator reduces to the already familiar version with just errors in the dependent variable. This method can be generalised to any differentiable function of x.

We can see that use of the function has saved us a fair amount of typing. A more relevant example (to physics) might be the function defined below. 

```python
import numpy as np

def quadraticRootFinder(a, b, c):
        
    if (b**2 - 4*a*c) < 0:
        print("Discriminant could not be calculated for ({0:.1f}x^2 + {1:.1f} + {2:.1f}).".format(a, b, c))
        solution1 = 0
        solution2 = 0
        return np.array([solution1, solution2])
    
    solution1 = (-b + np.sqrt(b**2 - 4*a*c))/(2*a)
    solution2 = (-b - np.sqrt(b**2 - 4*a*c))/(2*a)
    
    print("Roots of ({0:.1f}x^2 + {1:.1f}x + {2:.1f}): [{3:.3f}, {4:.3f}]".format(a, b, c, solution1, solution2))
    
    return np.array([solution1, solution2])
```

This function calculates the roots of a quadratic equation, $f(x) = ax^2 + bx + c$, taking the three parameters $(a, b, c)$ as inputs. The first section checks whether the roots are imaginary or not by attempting to evaluate the discriminant: if the inside of the square root is negative we don't have real roots! If this is okay, the function then uses the quadratic formula to evaluate the roots. Let's put this into practice.

```python
quadraticRootFinder(2, 4, 2)
quadraticRootFinder(2, 5, 2)
quadraticRootFinder(1, 2, 3)
```

This is a nice example, but what if we wanted to really take advantage of our function's functionality? How about calculating 500 pairs of roots with only 5 lines of code?

```python
for i in range(500):
    a = i*2 + 1;
    b = i*8 + 2*i;
    c = i*2 + 3*i;
    
    quadraticRootFinder(a, b, c)
```

This example uses arbitrary $(a, b, c)$ parameters. We could easily load a set of real parameters from a data file, and iterate over these. Another advantage with this approach is it doesn't matter if you want to calculate 500, 5000 or $5 \times 10^{10}$ parameters: the number of lines of code remain the same! (Of course, as you get to a large amount of iterations, your computing time increases. You only have to worry about this for very large numbers, however.)

## Using the $\chi^2$ function

Now we know all about functions, we can effectively use the $\chi^2$ functions that are given in the lab. These are detailed below.

```python
def fitFunc(p, x):
    '''
    Fit function
    '''
    f = p[0] + p[1]*x
    return f

def fitFuncDiff(p, x):
    '''
    Differential of fit function
    '''
    df = p[1]
    return df

def calcChiSq(p, x, y, xerr, yerr):
    '''
    Error function for fit
    '''
    e = (y - fitFunc(p, x))/(np.sqrt(yerr**2 + fitFuncDiff(p, x)**2*xerr**2))
    return e
```

The first function **fitFunc(p, x)** is our fit, or model, function. It takes an array of parameters, **p**, and our x-data values, **x**, as input. It returns y-values accordingly with the inputed parameters, and the x-values. In the example of a straight line model, there are two parameters: the intercept (denoted by **p[0]**) and the gradient (denoted by **p[1]**).

The second function, **fitFuncDiff(p, x)**, is similar to **fitFunc(p, x)**, but it uses the differential of the fit model instead of the fit model itself. This is used in calculating the $\chi^2$ with errors in both x & y, as detailed above.

The last function, **calcChiSq(p, x, y, xerr, yerr)**, returns an array of $\chi^2$ values, taking the fit parameters, the x and y data, and their errors as inputs.

To put it all into practice, let's generate some random data and fit it.

```python
import scipy.odr as odr
import matplotlib.pyplot as plt

y = fitFunc([2.5, 1.3], xdata)
y_noise = 0.2 * np.random.normal(size=xdata.size)

xdata = np.linspace(0, 4, 50)
ydata = y + y_noise

xerror = np.ones_like(xdata) * 0.10
yerror = np.ones_like(ydata) * 0.25

fig = plt.figure(figsize = (8, 6))
plt.errorbar(xdata, ydata, yerror, xerror, marker=".", linestyle = '')
plt.show


# Do fit
linearModel_1 = odr.Model(fitFunc)  
fitData_1 = odr.Data(xdata, ydata, xdata, ydata) 
odrObject_1 = odr.ODR(fitData_1, linearModel_1, beta0=[1.,1.])
odrFit_1 = odrObject_1.run()
odrFit_1.pprint() 

#Plot fit
plt.plot(xdata, fitFunc(odrFit_1.beta, xdata))
```

The fit looks pretty good, right? Don't worry if you didn't understand the random data generation or fit code above; there is another tutorial workbook which explains these methods in more detail. All you need to know is that we have some data and a fit. How good is the fit? Let's use the $\chi^2$ function to give a quantitative answer to this question.

```python
chi_array = calcChiSq(odrFit_1.beta, xdata, ydata, xerror, yerror)**2
chisqndf = np.sum(chi_array)/np.sqrt(len(xdata)-2)

print("ChiSq/NDF: {0:.3f}".format(chisqndf))
```

The above code is essentially all you need to run to get your $\chi^2$ value out of the functions. All we have done is provided our data, the errors in our data, and the fitted parameters. We then divide by the number of degrees of freedom ( (N-2), in this case: the -2 comes from the two fitted parameters in our straight line fit model). The final value is indicative of the quality of our fit. Remember: the closer to 1, the better.
